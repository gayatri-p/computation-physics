```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 1/np.sqrt(x**2 + 1)
def rectangle(f,a,b,n):
    h = (b-a)/(n-1)
    xs = a + np.arange(n)*h #np.linspace(a,b,n)
    fs = f(xs)
    plt.plot(xs, fs, 'Xb')
    real_xs = np.linspace(a,b,1000)
    real_fs = f(real_xs)
    plt.plot(real_xs, real_fs, 'r')

    # vertical lines
    for i in range(len(xs)):
        plt.vlines(xs[i], ymax=fs[i], ymin=min(fs), color='g')
        plt.vlines(xs[i]+h, ymax=fs[i], ymin=min(fs), color='g')
        if i != len(xs) - 1:
            plt.hlines(fs[i], xmin=xs[i], xmax=xs[i]+h, color='g')
        # print(xs[i], fs[i])
    # plt.vlines(xs[-1]+h, ymax=fs[-1], ymin=min(fs))
    return h*np.sum(fs)

ans = np.log(1 + np.sqrt(2))
print(ans)
rectangle(f, 0., 1, 11)
```

```python
def trapezoid(f,a,b,n):
    h = (b-a)/(n-1)
    xs = a + np.arange(n)*h
    cs = np.ones(n); cs[0] = 0.5; cs[-1] = 0.5
    contribs = cs*f(xs)
    return h*np.sum(contribs)
def simpson(f,a,b,n):
    h = (b-a)/(n-1)
    xs = a + np.arange(n)*h
    cs = 2*np.ones(n)
    cs[1::2] = 4; cs[0] = 1; cs[-1] = 1
    contribs = cs*f(xs)
    return (h/3)*np.sum(contribs)
```

```python
ans = np.log(1 + np.sqrt(2))
print(ans)
for integrator in (rectangle, trapezoid, simpson):
    print(integrator(f, 0., 1., 51), end=" ")
```
