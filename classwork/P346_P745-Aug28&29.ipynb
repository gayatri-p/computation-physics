Solving $Ax=b$ for lower and upper triangular matrices

```python
import numpy as np
def forsub(L,bs):
  n = bs.size
  xs = np.zeros(n)
  for i in range(n):
    xs[i] = (bs[i] - L[i,:i]@xs[:i])/L[i,i]
  return xs
def backsub(U,bs):
  n = bs.size
  xs = np.zeros(n)
  for i in reversed(range(n)):
    xs[i] = (bs[i] - U[i,i+1:]@xs[i+1:])/U[i,i]
  return xs
def testcreate(n,val):
  A = np.arange(val,val+n*n).reshape(n,n)
  A = np.sqrt(A)
  bs = (A[0,:])**2.1
  return A, bs
def testsolve(f,A,bs):
  xs = f(A,bs); print(xs)
  xs = np.linalg.solve(A,bs); print(xs)

A, bs = testcreate(4,21)
print(A)
print(bs)
L = np.tril(A)
print(L)
testsolve(forsub,L,bs)
print(" ")
U = np.triu(A)
print(U)
testsolve(backsub,U,bs)
```

Gaussian elimination


```python
import numpy as np
def gauelim(inA,inbs):
  A = np.copy(inA)
  bs = np.copy(inbs)
  n = bs.size
  for j in range(n-1):
    for i in range(j+1,n):
      coeff = A[i,j]/A[j,j]
      A[i,j:] -= coeff*A[j,j:]
      bs[i] -= coeff*bs[j]
  xs = backsub(A,bs)
  return xs

def forsub(L,bs):
  n = bs.size
  xs = np.zeros(n)
  for i in range(n):
    xs[i] = (bs[i] - L[i,:i]@xs[:i])/L[i,i]
  return xs
def backsub(U,bs):
  n = bs.size
  xs = np.zeros(n)
  for i in reversed(range(n)):
    xs[i] = (bs[i] - U[i,i+1:]@xs[i+1:])/U[i,i]
  return xs
def testcreate(n,val):
  A = np.arange(val,val+n*n).reshape(n,n)
  A = np.sqrt(A)
  bs = (A[0,:])**2.1
  return A, bs
def testsolve(f,A,bs):
  xs = f(A,bs); print(xs)
  xs = np.linalg.solve(A,bs); print(xs)
#if _name == '_main_':
A, bs = testcreate(4,21)
print(A)
print(bs)
testsolve(gauelim,A,bs)
```
